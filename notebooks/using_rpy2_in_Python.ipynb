````python
import rpy2

import pandas as pd

import warnings
warnings.filterwarnings('ignore')

from rpy2 import robjects as ro
from rpy2.robjects import pandas2ri, rl
from rpy2.robjects.packages import importr

pandas2ri.activate()

R = ro.r

dplyr = importr('dplyr')
stats = importr('stats')
base = importr('base')
````

````python
affairs_df = pd.read_csv("https://raw.githubusercontent.com/mark-andrews/intropy4ds/master/data/affairs.csv")
````

```
dplyr::mutate(affairs_df, had_affair = affairs > 0) 
```

````python
affairs_df = dplyr.mutate(affairs_df, had_affair = rl('affairs > 0'))
````

````python
affairs_df
````

```
M <- glm(had_affair ~ age + gender + yearsmarried, 
         data = affairs_df, 
         family = binomial(link = 'logit'))
```

````python
M = stats.glm("had_affair ~ age + gender + yearsmarried",
              data = affairs_df,
              family = R("binomial(link = 'logit')"))
````

````python
S = base.summary(M)
````

````python
print(S)
````

````python
S.rx2('coefficients').round(3)
````
